```python
df = pd.read_excel(f_path)
```

```python
d = {'swing': 0, 'impact': 1, 'castoff': 2}
df['name'] = df['name'].map(d)
```

```python
features = ['Shape', 'Eccentricity', 'Angle', 'linear','Convex']
X = df[features]
y = df['name']

print(X)
print(y)
```

```python
X=df.iloc[:,:-1]
```

```python
X
```

```python
y=df.iloc[:,5]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

DecisionTree 정확도

```python
from sklearn import tree
dt_clf = tree.DecisionTreeClassifier(max_depth=10)
dt_clf.fit(X_train, y_train)


dt_clf.score(X_test, y_test)

y_pred= dt_clf.predict(X_test)
count = 0
print(len(y_test))
for i, data in enumerate(y_test):
  print(data)
  if data == y_pred[i]:
    count += 1
print(count)
print('Acc:', count/len(y_test))


dt_clf.score(X_test, y_test)
```

RandomForestClassifier

```python
from sklearn import ensemble

n_estimators = 100
max_depth = 10
for i in range(1, n_estimators+1):
  for j in range(1, max_depth+1):
    rf_clf= ensemble.RandomForestClassifier(n_estimators=i, max_depth=j)
    rf_clf.fit(X_train, y_train)
    y_pred= rf_clf.predict(X_test)
    count = 0
    for index, data in enumerate(y_test):
      if data == y_pred[index]:
        count += 1
    print('[n_estimators={0}, max_depth={1}] Acc: {2}'.format(i, j, count/len(y_test)))
    

#rf_clf.score(X_test, y_test)
```

```python
from sklearn import ensemble
rf_clf= ensemble.RandomForestClassifier(n_estimators=47, max_depth=8, min_samples_split =6, min_samples_leaf =4,class_weight ='balanced' ,n_jobs=-1, 
                           random_state=2, oob_score=True)
rf_clf.fit(X_train, y_train)
rf_clf.score(X_test, y_test)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

# feature importance 추출 
print("Feature importances:\n{0}".format(np.round(rf_clf.feature_importances_, 3)))

# feature별 importance 매핑
for name, value in zip(features , rf_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))

# feature importance를 column 별로 시각화 하기 
#sns.barplot(x=rf_clf.feature_importances_ , y=features)
#plt.bar(features,rf_clf.feature_importances_ ,color = ['r','g','b','y','m'])
plt.barh(features,rf_clf.feature_importances_ ,color = ['orange','orangered','forestgreen','royalblue','blueviolet'])


plt.savefig('Feature importances.eps', format='eps')
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
preds = rf_clf.predict(X_test)
confusion_matrix(y_test,preds)
plot_confusion_matrix(rf_clf,X_test,y_test)
```
